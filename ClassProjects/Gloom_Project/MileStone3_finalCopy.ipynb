```python
import networkx as nx
import matplotlib.pyplot as plt
import pytumblr
```

```python
lotr = nx.read_graphml('lotr.graphml')
```

```python
#BLOGS for LOTR unfiltered
lotr_nodes_dict = dict(nx.get_node_attributes(lotr, 'type'))
blogs = []

for node in lotr_nodes_dict:
    if lotr_nodes_dict[node] == 'blog':
        blogs.append(node)

lotr_projected = nx.bipartite.weighted_projected_graph(lotr, blogs)
nx.draw(lotr_projected, with_labels=True)
nx.write_graphml(lotr_projected, 'lotrBlogs.graphml')
```

```python
#we will have to filter in the future due to lack of data!!! But these are the computations we need
lotr_filtered = nx.Graph(lotr_projected)
to_remove = [(n1, n2) for n1, n2, data in lotr_projected.edges(data=True) if data['weight'] <= 1]
lotr_filtered.remove_edges_from(to_remove)
lotr_projected2 = nx.bipartite.weighted_projected_graph(lotr_filtered, blogs)
nx.draw(lotr_projected2, with_labels=True)
```

```python
#TAGS for LOTR unflitered
lotr_nodes_dict2 = dict(nx.get_node_attributes(lotr, 'type'))
tags = []

for node in lotr_nodes_dict2:
    if lotr_nodes_dict2[node] == 'tag':
        tags.append(node)

lotr_projected3 = nx.bipartite.weighted_projected_graph(lotr, tags)
nx.draw(lotr_projected3, with_labels=True)
nx.write_graphml(lotr_projected3, 'lotrTags.graphml')
```

```python
#we will have to filter in the future due to lack of data!!! But these are the computations we need
lotr_filtered2 = nx.Graph(lotr_projected3)
to_remove = [(n1, n2) for n1, n2, data in lotr_projected3.edges(data=True) if data['weight'] <= 1]
lotr_filtered2.remove_edges_from(to_remove)
lotr_projected4 = nx.bipartite.weighted_projected_graph(lotr_filtered2, tags)
nx.draw(lotr_projected4, with_labels=True)
```
